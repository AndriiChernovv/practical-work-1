5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
# Клас 3D-фігура
class Shape3D:
    def get_volume(self):
        """Метод для отримання об'єму фігури"""
        return 0

    def __add__(self, other):
        """Метод для складання об'ємів двох 3D-фігур"""
        if isinstance(other, Shape3D):
            return self.get_volume() + other.get_volume()
        raise TypeError("Можна додавати лише екземпляри класу Shape3D")

# Приклад
shape1 = Shape3D()
shape2 = Shape3D()
print(shape1 + shape2)  # Виведе 0
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math # взяв бібліотеку для числа пі
# Дочірній клас Куб
class Cube(Shape3D):
    def __init__(self, side_length):
        self.side_length = side_length

    def get_volume(self):
        """Розрахунок об'єму куба"""
        return self.side_length ** 3

# Дочірній клас Циліндр
class Cylinder(Shape3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def get_volume(self):
        """Розрахунок об'єму циліндра"""
        return math.pi * (self.radius ** 2) * self.height

cube = Cube(3)
cylinder = Cylinder(2, 5)

# Обчислення загального об'єму
print(f"Об'єм куба: {cube.get_volume()}")
print(f"Об'єм циліндра: {cylinder.get_volume()}")
print(f"Сумарний об'єм: {cube + cylinder}")
```

7. (3б). Продемонструйте різницю між isinstance та issubclass.

```python
print(isinstance(cube, Shape3D))  # True, оскільки cube є екземпляром Shape3D
print(isinstance(cylinder, Cube))  # False, оскільки cylinder не є екземпляром Cube
print(issubclass(Cube, Shape3D))  # True, оскільки Cube є підкласом Shape3D
print(issubclass(Shape3D, Cube))  # False, оскільки Shape3D не є підкласом Cube
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, wallet_id, balance=0.0):
        self.__id = wallet_id  # Приватний ідентифікатор гаманця
        self.__balance = balance  # Приватний баланс

    @property
    def balance(self):
        """Отримання балансу"""
        return self.__balance

    @balance.setter
    def balance(self, amount):
        """Забороняємо прямий запис балансу"""
        raise AttributeError("Зміна балансу напряму заборонена. Використовуйте методи поповнення або зняття коштів.")

    def deposit(self, amount):
        """Поповнення балансу"""
        if amount > 0:
            self.__balance += amount
            return f"Гаманець {self.__id} поповнено на {amount}. Новий баланс: {self.__balance}"
        return "Сума поповнення має бути більшою за 0."

    def withdraw(self, amount):
        """Зняття коштів"""
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"З гаманця {self.__id} знято {amount}. Новий баланс: {self.__balance}"
        return "Недостатньо коштів або некоректна сума."

# Приклад
wallet = EWallet("Wallet_1234", 100.0)
print(wallet.deposit(50))  # Поповнення на 50
print(wallet.withdraw(30))  # Зняття 30
print(wallet.balance)  # Отримання балансу
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name}: {self.price} грн"

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        """Додає товар до кошика"""
        self.items.append(item)
        return f"Товар {item.name} додано до кошика."

    def remove_item(self, item_name):
        """Видаляє товар з кошика за ім'ям"""
        for item in self.items:
            if item.name == item_name:
                self.items.remove(item)
                return f"Товар {item_name} видалено з кошика."
        return "Товар не знайдено у кошику."

    def display_items(self):
        """Відображає список товарів у кошику"""
        if not self.items:
            return "Кошик порожній."
        return "\n".join(str(item) for item in self.items)

# Приклад
cart = ShoppingCart()
item1 = Item("Яблуко", 15)
item2 = Item("Молоко", 35)

print(cart.add_item(item1))
print(cart.add_item(item2))
print(cart.display_items())
print(cart.remove_item("Яблуко"))
print(cart.display_items())
```

```python
# Створення рядка з 4 різними символами
string = "abcd"

# Створення ітератора
iterator = iter(string)

# Виклик next() 5 разів
print(next(iterator))  # 'a'
print(next(iterator))  # 'b'
print(next(iterator))  # 'c'
print(next(iterator))  # 'd'

# П'ята спроба викличе StopIteration
try:
    print(next(iterator))
except StopIteration:
    print("Ітерація завершена: більше елементів немає.")
```

Ітерабельні об'єкти, такі як рядки та списки, містять дані та можуть бути перетворені на ітератор(як в попередньому завданні), тоді як ітератор вже містить механізм поступового отримання елементів, використовуючи функцію next()

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class DaysOfWeekIterator:
    def __init__(self):
        self.days = ["Понеділок", "Вівторок", "Середа", "Четвер", "П'ятниця", "Субота", "Неділя"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        raise StopIteration

# Створення ітератора
week_iterator = DaysOfWeekIterator()

# Виклик next() для кожного дня тижня
print(next(week_iterator))  # Понеділок
print(next(week_iterator))  # Вівторок
print(next(week_iterator))  # Середа
print(next(week_iterator))  # Четвер
print(next(week_iterator))  # П'ятниця
print(next(week_iterator))  # Субота
print(next(week_iterator))  # Неділя

# Вихід за межі списку викличе StopIteration
try:
    print(next(week_iterator))
except StopIteration:
    print("Ітерація завершена: більше днів немає.")
```
